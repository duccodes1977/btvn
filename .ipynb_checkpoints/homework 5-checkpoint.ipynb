```python
# W6A16
def WxA16(matrix):
    for row in matrix:
        print("".join(f"{x:4d}" for x in row))



```

```python
# W6A17
def WxA17(a):
    n = len(a)
    print(*[a[i][i] for i in range(n)])
    print(*[a[i][n-1-i] for i in range(n)])


```

```python
# W6A18
def WxA18(a, k):
    print(sum(row[k] for row in a), sum(a[k]))


```

```python
# W6A19
def WxA19(lst):
    idx = 0
    for i in range(len(lst)):
        if lst[i] > lst[idx]:
            idx = i
    return idx


```

```python
# W6A20
def WxA20(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1


```

```python
# W6A21
def WxA21(line):
    words = line.split()
    d = {}
    for i in range(len(words)):
        d.setdefault(words[i], []).append(i)
    return {k: tuple(v) for k, v in d.items()}


```

```python
# W6A22
def WxA22(lst):
    idx = 0
    for i in range(len(lst)):
        if lst[i] > lst[idx]:
            idx = i
    return idx


```

```python
# W6A23
def WxA23(lst, x):
    return [v for v in lst if v >= x]


```

```python
# W6A24
def WxA24(lst):
    return [0]*lst.count(0) + [v for v in lst if v != 0]


```

```python
# W6A25
def WxA25(lst):
    return [i for i in range(10) if i not in lst]


```

```python
# W6A26
def WxA26(lst):
    s = 0
    for i in range(1, len(lst)-1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            s += lst[i]
    return s


```

```python
# W6A27
def WxA27(lst, x):
    res = [i for i in range(len(lst)) if lst[i] < x]
    return res if res else [-1]


```

```python
# W6A28
def WxA28(lst):
    inc = True
    dec = True
    for i in range(len(lst)-1):
        if lst[i] >= lst[i+1]:
            inc = False
        if lst[i] <= lst[i+1]:
            dec = False
    if inc:
        return "tăng thực sự"
    if dec:
        return "giảm thực sự"
    return "vô danh"
```
