```python
#W9A1

def merge(a, b):
    i = j = 0
    res = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            res.append(a[i])
            i += 1
        else:
            res.append(b[j])
            j += 1
    res.extend(a[i:])
    res.extend(b[j:])
    return res
```

```python
#W9A2

import heapq

def kth_largest(matrix, k):
    n, m = len(matrix), len(matrix[0])
    heap = []
    for i in range(n):
        for j in range(m):
            heapq.heappush(heap, matrix[i][j])
            if len(heap) > k:
                heapq.heappop(heap)
    return heap[0]
```

```python
#W9A3
def smallest_by_swap(num):
    s = list(num)
    n = len(s)
    for i in range(n):
        min_pos = i
        for j in range(i+1, n):
            if s[j] < s[min_pos]:
                min_pos = j
        if min_pos != i and s[min_pos] != '0':
            s[i], s[min_pos] = s[min_pos], s[i]
            break
    return ''.join(s)
```

```python
#W9A4
def can_convert(a, b):
    n, m = len(a), len(b)
    dp = [[False]*(m+1) for _ in range(n+1)]
    dp[0][0] = True

    for i in range(1, n+1):
        if a[i-1].islower():
            dp[i][0] = dp[i-1][0]

    for i in range(1, n+1):
        for j in range(1, m+1):
            if a[i-1].upper() == b[j-1]:
                dp[i][j] = dp[i-1][j-1]
            if a[i-1].islower():
                dp[i][j] |= dp[i-1][j]

    return "YES" if dp[n][m] else "NO"
```

```python
#W9A5
def is_subarray(a, b):
    n, m = len(a), len(b)
    for i in range(m - n + 1):
        if b[i:i+n] == a:
            return "YES"
    return "NO"
```

```python
#W9A6

def next_permutation(a):
    n = len(a)
    i = n - 2
    while i >= 0 and a[i] >= a[i+1]:
        i -= 1
    if i == -1:
        return a[::-1]

    j = n - 1
    while a[j] <= a[i]:
        j -= 1

    a[i], a[j] = a[j], a[i]
    a[i+1:] = reversed(a[i+1:])
    return a
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    i = 0
    while i < len(flowerbed):
        if flowerbed[i] == 0:
            left = (i == 0 or flowerbed[i-1] == 0)
            right = (i == len(flowerbed)-1 or flowerbed[i+1] == 0)
            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True
                i += 2
                continue
        i += 1
    return k <= 0
```

```python
#W9A8

def is_path_crossing(moves):
    x = y = 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'R': x += 1
        elif move == 'L': x -= 1
        elif move == 'U': y += 1
        elif move == 'D': y -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
```

```python
#W9A9

def Median(a):
    x1, x2, x3, x4, x5 = a

    if x1 > x2: x1, x2 = x2, x1
    if x3 > x4: x3, x4 = x4, x3
    if x1 > x3: x1, x3 = x3, x1
    if x2 > x4: x2, x4 = x4, x2
    if x2 > x3: x2, x3 = x3, x2

    return x3 if x3 < x5 else max(x2, x5)
```

```python
#W9A10

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    res = 0
    j = 0
    for h in houses:
        while j+1 < len(heaters) and abs(heaters[j+1]-h) <= abs(heaters[j]-h):
            j += 1
        res = max(res, abs(heaters[j]-h))
    return res
```
