```python
a = int(input())

i = 1
while i <= 10:
    print(f"{a} x {i} = {a*i}")
    i += 1
```

```python
h = int(input())

i = 1
while i <= h:
    j = 1
    while j <= h - i:
        print(" ", end="")
        j += 1

    j = 1
    while j <= 2*i - 1:
        if j == 1 or j == 2*i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="")
        j += 1

    print()
    i += 1
```

```python
n = int(input())

tong = 0
i = 1
while i <= n:
    tong += i
    i += 1

print(tong)
```

```python
a = int(input())
b = int(input())

i = 1
while i <= min(a, b):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
    i += 1
```

```python
a = int(input())

i = 0
while i*i <= a:
    if i*i == a:
        print("La so chinh phuong")
        break
    i += 1
else:
    print("Khong phai so chinh phuong")
```

```python
a = int(input())

tong = 0
i = 1
while i < a:
    if a % i == 0:
        tong += i
    i += 1

if tong == a:
    print("La so hoan hao")
else:
    print("Khong phai so hoan hao")
```

```python
n = int(input())

tong = 0
k = 0

while tong + k + 1 < n:
    k += 1
    tong += k

print(k)
```

```python
A = float(input())

tong = 0
n = 0

while tong <= A:
    n += 1
    tong += 1/n

print(n)
```

```python
first = True

while True:
    x = int(input())
    if x == -1:
        break

    if first:
        max_val = x
        min_val = x
        first = False
    else:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x

print("Max:", max_val)
print("Min:", min_val)
```
