```python
def W7A16(arr):
    from collections import Counter

    freq = Counter(arr)
    delete_count = 0

    for x, cnt in freq.items():
        if cnt != x:
            delete_count += abs(cnt - x)

    if delete_count == 0:
        print("YES")
    else:
        print("NO", delete_count)
```

```python
def W7A17(arr):
    cand, count = None, 0
    for x in arr:
        if count == 0:
            cand, count = x, 1
        elif x == cand:
            count += 1
        else:
            count -= 1
    if arr.count(cand) > len(arr)//2:
        print(cand)
```

```python
def W7A18(arr):
    even = sorted([x for x in arr if x % 2 == 0])
    odd = sorted([x for x in arr if x % 2 == 1], reverse=True)
    print(*(even + odd))
```

```python
def W7A19(a, b):
    i = j = 0
    res = []
    inc = a[0] <= a[-1]

    while i < len(a) and j < len(b):
        if (a[i] <= b[j]) == inc:
            res.append(a[i]); i += 1
        else:
            res.append(b[j]); j += 1
    res.extend(a[i:])
    res.extend(b[j:])
    print(*res)
```

```python
def W7A20(arr, k):
    s = sum(x for x in arr if len(str(abs(x))) == k)
    print(s)
```

```python
def W7A21(d, m):
    days = [31,29,31,30,31,30,31,31,30,31,30,31]
    if m < 1 or m > 12 or d < 1 or d > days[m-1]:
        print("invalid"); return
    print(sum(days[:m-1]) + d)
```

```python
from collections import Counter

def W7A22(names):
    first_names = [n.split()[-1] for n in names]
    cnt = Counter(first_names)
    mx = max(cnt.values())
    print(min(k for k,v in cnt.items() if v == mx))
```

```python
import re

def W7A23(s):
    names = re.findall(r'\b[A-Z][a-z]{1,}\b', s)
    print(len(names))

```

```python
def W7A24(s):
    words = s.split()
    res = [words[0]]
    for w in words[1:]:
        if w != res[-1]:
            res.append(w)
    print(" ".join(res))
```

```python
def W7A25(num, k):
    stack = []
    remove = len(num) - k
    for d in num:
        while stack and remove and stack[-1] < d:
            stack.pop(); remove -= 1
        stack.append(d)
    print("".join(stack[:k]))
```

```python
import re

def W7A26(s):
    nums = list(map(int, re.findall(r'\d+', s)))
    total = sum(nums)
    avg = total / len(nums)
    print(total, round(avg,2))
```

```python
def W7A27(s):
    stack = []
    mp = {')':'(',']':'[','}':'{'}
    for c in s:
        if c in '([{': stack.append(c)
        elif c in mp:
            if not stack or stack.pop() != mp[c]:
                print("NO"); return
    print("YES" if not stack else "NO")
```

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0: return False
    return True

def W7A28(N):
    best, cur = [], []
    for i in range(4, N+1):
        if not is_prime(i):
            cur.append(i)
        else:
            if len(cur) > len(best): best = cur
            cur = []
    if len(cur) > len(best): best = cur
    print(*best)
```

```python
def W7A29(arr):
    best = cur = [arr[0]]
    for i in range(1,len(arr)):
        if arr[i] > arr[i-1]:
            cur.append(arr[i])
        else:
            if len(cur) > len(best): best = cur
            cur = [arr[i]]
    if len(cur) > len(best): best = cur
    print(*best)
    
```

```python
import bisect

def W7A30(arr):
    dp = []
    for x in arr:
        i = bisect.bisect_left(dp, x)
        if i == len(dp): dp.append(x)
        else: dp[i] = x
    print(len(dp))
```

```python
def W7A31(s):
    s = list(s)
    for i in range(len(s)):
        mn = min(s[i+1:])
        if mn < s[i]:
            j = ''.join(s).rfind(mn)
            s[i], s[j] = s[j], s[i]
            break
    print("".join(s))
```
