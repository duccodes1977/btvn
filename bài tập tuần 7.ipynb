```python
def W7A1(arr, x):
    l, r = 0, len(arr) - 1
    while l <= r:
        m = (l + r) // 2
        if arr[m] == x:
            return m
        if arr[m] < x:
            l = m + 1
        else:
            r = m - 1
    return -1
```

```python
def W7A2(arr, x):
    return arr.count(x)
```

```python
def W7A3(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a
```

```python
def W7A4(arr):
    max_count = 0
    result = arr[0]
    for x in arr:
        c = arr.count(x)
        if c > max_count:
            max_count = c
            result = x
    return result
```

```python
def W7A5(arr, x):
    cnt = 0
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == x:
                cnt += 1
    return cnt
```

```python
def W7A6(arr):
    res = []
    for sub in arr:
        for x in sub:
            res.append(x)
    return res
```

```python
def W7A7(arr):
    n = len(arr)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if arr[j] < arr[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
```

```python
def W7A8(intervals, queries):
    res = []
    for q in queries:
        best = -1
        for l, r in intervals:
            if l <= q <= r:
                length = r - l + 1
                if best == -1 or length < best:
                    best = length
        res.append(best)
    return res
```

```python
def W7A9(arr):
    res = []
    for i in range(len(arr)):
        cnt = 0
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[i]:
                cnt += 1
        res.append(cnt)
    return res
```

```python
def W7A10(arr, m):
    max_sum = 0
    for i in range(len(arr)):
        s = 0
        for j in range(i, len(arr)):
            s = (s + arr[j]) % m
            max_sum = max(max_sum, s)
    return max_sum
```

```python
def W7A11(sentence, word):
    return sentence.split().count(word)
```

```python
def W7A12(a, b, c, d):
    arr = [('A', a), ('B', b), ('C', c), ('D', d)]
    arr.sort(key=lambda x: x[1])
    return [x[0] for x in arr]
```

```python
def W7A13(arr):
    from collections import Counter
    cnt = Counter(arr)
    ans = 0
    for x in cnt:
        if x + 1 in cnt:
            ans = max(ans, cnt[x] + cnt[x + 1])
    return ans
```

```python
def W7A14(arr):
    pos = [i for i in range(len(arr)) if arr[i] == 7]
    if not pos:
        return "Not found"
    return sorted(pos, reverse=True)
```
