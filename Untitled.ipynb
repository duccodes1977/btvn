```python
class Robot:
    def __init__(self, name, hp, atk):
        self.name = name
        self.hp = hp
        self.atk = atk

    def is_alive(self):
        return self.hp > 0

    def receive_damage(self, amount):
        self.hp -= amount
        if self.hp < 0:
            self.hp = 0

    def attack(self, other):
        other.receive_damage(self.atk)


class Tanker(Robot):
    def __init__(self, name, hp, atk, defense):
        super().__init__(name, hp, atk)
        self.defense = defense
        self.shield = 0

    def skillFortify(self):
        self.shield += 1

    def receive_damage(self, amount):
        damage = amount - self.defense - self.shield
        if damage < 0:
            damage = 0
        self.hp -= damage
        if self.hp < 0:
            self.hp = 0
        self.shield = 0


class Assassin(Robot):
    def __init__(self, name, hp, atk):
        super().__init__(name, hp, atk)
        self.crit_count = 0
        self.stealth = False

    def attack(self, other):
        if self.crit_count == 3:
            other.receive_damage(self.atk * 2)
            self.crit_count = 0
        else:
            other.receive_damage(self.atk)
            self.crit_count += 1

    def receive_damage(self, amount):
        if self.stealth:
            self.hp -= amount / 2
        else:
            self.hp -= amount
        if self.hp < 0:
            self.hp = 0

    def noi_tai(self):
        if self.hp < 30:
            self.stealth = True


# ===== MAIN =====

# nhập: hp atk defense
t_hp, t_atk, t_def = map(float, input().split())
# nhập: hp atk
a_hp, a_atk = map(float, input().split())

n = int(input())

tank = Tanker("T1", t_hp, t_atk, t_def)
assassin = Assassin("A1", a_hp, a_atk)

for i in range(n):
    if not tank.is_alive() or not assassin.is_alive():
        break

    assassin.attack(tank)
    tank.attack(assassin)
    assassin.noi_tai()

print("T1: HP =", tank.hp)
print("A1: HP =", assassin.hp)
```
