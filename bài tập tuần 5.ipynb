```python
def Max_Of_Two(a, b):
    if a > b:
        return a
    return b
```

```python
def swap(a, b):
    return b, a
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def is_perfect(n):
    if n <= 1:
        return False
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
```

```python
def first_index(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
```

```python
def factorial(n):
    gt = 1
    for i in range(1, n+1):
        gt *= i
    return gt
```

```python
def calculator(num1, num2, operat):
    if operat == '+':
        kq = num1 + num2
    elif operat == '-':
        kq = num1 - num2
    elif operat == '*':
        kq = num1 * num2
    elif operat == '/':
        kq = num1 / num2
    else:
        return None
    return round(kq, 2)
```

```python
def hamming_distance(x, y):
    b1 = bin(x)[2:]
    b2 = bin(y)[2:]
    max_len = max(len(b1), len(b2))
    b1 = b1.zfill(max_len)
    b2 = b2.zfill(max_len)

    dem = 0
    for i in range(max_len):
        if b1[i] != b2[i]:
            dem += 1
    return dem
```

```python
def sum_digits(n):
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
```

```python
def is_isomorphic(a, b):
    map1 = {}
    map2 = {}

    for i in range(len(a)):
        c1 = a[i]
        c2 = b[i]

        if c1 in map1 and map1[c1] != c2:
            return False
        if c2 in map2 and map2[c2] != c1:
            return False

        map1[c1] = c2
        map2[c2] = c1

    return True
```
