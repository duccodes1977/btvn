```python
import math

a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))

if a + b <= c or a + c <= b or b + c <= a:
    print("Khong la ba canh cua mot tam giac")
else:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")

    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))

    print("Chu vi:", chu_vi)
    print("Dien tich:", dien_tich)
```

```python
import math

a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))

if a == 0:
    print("Khong phai phuong trinh bac 2")
else:
    delta = b*b - 4*a*c
    if delta < 0:
        print("Phuong trinh vo nghiem")
    elif delta == 0:
        x = -b / (2*a)
        print("Nghiem kep x =", x)
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("x1 =", x1)
        print("x2 =", x2)

```

```python
a = float(input("a: "))
b = float(input("b: "))
c = float(input("c: "))
d = float(input("d: "))
m = float(input("m: "))
n = float(input("n: "))

D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c

if D == 0:
    if Dx == 0 and Dy == 0:
        print("He vo so nghiem")
    else:
        print("He vo nghiem")
else:
    x = Dx / D
    y = Dy / D
    print("x =", x)
    print("y =", y)
```
